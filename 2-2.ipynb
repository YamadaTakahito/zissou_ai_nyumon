```python
from sklearn.model_selection import train_test_split
from mglearn.datasets import load_extended_boston
```

```python
X, y = load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model

X_train_single = X_train[:, 5].reshape(-1, 1)
X_test_single = X_test[:, 5].reshape(-1, 1)

lm_single = linear_model.LinearRegression()
lm_single.fit(X_train_single, y_train)

y_pred_train = lm_single.predict(X_test_single)

plt.xlabel("RM")
plt.ylabel("MEDV")
plt.scatter(X_train_single, y_train)
plt.plot(X_train_single, y_pred_train, color="red", linewidth=2)
plt.show()
```

```python
print(f"intercept: {lm_single.intercept_}")
print(f"coef: {lm_single.coef_}")

print(f"Train score: {lm_single.score(X_train_single, y_train)}")
print(f"Test score: {lm_single.score(X_test_single[:127, :], y_test)}")
```

```python
lm = linear_model.LinearRegression()
lm.fit(X_train, y_train)

print(f"intercept: {lm.intercept_}")
print(f"coef: {lm.coef_}")

print(f"Train score: {lm.score(X_train, y_train)}")
print(f"Test score: {lm.score(X_test, y_test)}")
```

```python
ridge = linear_model.Ridge(alpha=1.0, random_state=0)
ridge.fit(X_train, y_train)

print(f"intercept: {ridge.intercept_}")
print(f"coef: {ridge.coef_}")

print(f"Train score: {ridge.score(X_train, y_train)}")
print(f"Test score: {ridge.score(X_test, y_test)}")
```

```python
import numpy as np
lasso = linear_model.Lasso(alpha=0.01, max_iter=2000, random_state=0)
lasso.fit(X_train, y_train)

print(f"intercept: {lasso.intercept_}")
print(f"coef: {lasso.coef_}")
print(f"Number of features used: {np.sum(lasso.coef_ != 0)}")
print(f"Train score: {lasso.score(X_train, y_train)}")
print(f"Test score: {lasso.score(X_test, y_test)}")
```
